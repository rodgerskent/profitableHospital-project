```python
import pandas as pd
```

```python
#2017 Hospital Data from Centers for Medicare Services
year2017 = pd.read_csv("main4_2017.csv")
year2017.head(2)
```

```python
#2016 Hospital Data from Centers for Medicare Services
year2016 = pd.read_csv("main4_2016.csv")
year2016.head(2)
```

```python
#2015 Hospital Data from Centers for Medicare Services
year2015 = pd.read_csv("main4_2015.csv")
year2015.head(2)
```

```python
#Lattitude & Longitude for the 2017 hospital listing
geoTable = pd.read_csv("geocoded2017.csv")
geoTable
```

```python
#Append a lattitude & longitude
tableau2017 = pd.merge(year2017, geoTable, how= "left", on=['Provider CCN'])
tableau2017.head(2)
```

```python
tableau2017.to_csv("tableau2017.csv", index=False, header=True)
```

```python
#Append a lattitude & longitude
tableau2016 = pd.merge(year2016, geoTable, how= "left", on=['Provider CCN'])
tableau2016.head(2)
```

```python
tableau2016.to_csv("tableau2016.csv", index=False, header=True)
```

```python
#Append a lattitude & longitude
tableau2015 = pd.merge(year2015, geoTable, how= "left", on=['Provider CCN'])
tableau2015.head(2)
```

```python
tableau2015.to_csv("tableau2015.csv", index=False, header=True)
```

```python
#Merge 2017 and 2016
df = pd.merge(year2017, year2016, how= "outer", on=['Provider CCN'])
df.head(2)
```

```python
#Merge 2016-2017 with 2015
df2 = pd.merge(df, year2015, how= "outer", on=['Provider CCN'])
df2.head(2)
```

```python
df2.to_csv("XXmain4tableau.csv", index=False, header=True)
```
