```python
import pandas as pd
```

```python
#2017 Hospital Data from Centers for Medicare Services
df = pd.read_csv("mainFile4.csv")
df.head(2)
```

```python
#Lattitude & Longitude for the 2017 hospital listing
geoTable = pd.read_csv("geocoded2017.csv")
geoTable
```

```python
geoTable["Provider CCN"].dtype
```

```python
geoTable["lat"].dtype
```

```python
geoTable["lng"].dtype
```

```python
geoTable["Provider CCN"] = pd.to_numeric(geoTable["Provider CCN"])
```

```python
geoTable["Provider CCN"].dtype
```

```python
geoTable["Provider CCN"] = geoTable["Provider CCN"].apply(pd.to_numeric, errors='coerce')
```

```python
geoTable["Provider CCN"].dtype
```

```python
#Append a lattitude & longitude
tabFile1 = pd.merge(df, geoTable, how= "left", on=['Provider CCN'])
tabFile1.head(2)
```

```python
tabFile1["Hospital Name"].drop_duplicates
tabFile1.head()
```

```python
tabFile1.to_csv("main4tableau.csv", index=False, header=True)
```
